## The Loop Pattern

Every loop has 4 phases:

```
  plan → execute → evaluate → decide
    ↑                            ↓
    └────────────────────────────┘
```

- **plan**: What should we try next?
- **execute**: Run the experiment
- **evaluate**: How did it do?
- **decide**: Ship, retrain, or pivot?

Let's implement a minimal loop: count until we hit a target.

````python
import sys
sys.path.insert(0, '.')

from loops.base import Loop, Decision, Experiment

class CountingLoop(Loop):
    """Count until we hit the target."""
    
    def __init__(self, target=5):
        super().__init__("counting")
        self.target = target
        self.count = 0
    
    def plan(self, context):
        return {"action": "increment", "current": self.count, "target": self.target}
    
    def execute(self, plan):
        self.count += 1
        return {"count": self.count, "target": self.target}
    
    def evaluate(self, results):
        reached = results["count"] >= results["target"]
        return {"reached_target": reached, "progress": f"{results['count']}/{results['target']}"}
    
    def decide(self, metrics, history):
        if metrics["reached_target"]:
            return Decision.SHIP, f"Reached {self.target}!"
        return Decision.CONTINUE, f"{metrics['progress']} — keep going"

# Run it!
loop = CountingLoop(target=5)
loop.run_until(Decision.SHIP, max_iterations=10)

print(f"\nDone! {len(loop.history)} iterations:")
for exp in loop.history:
    print(f"  {exp.id}: count={exp.results['count']} → {exp.decision.value}")
````
